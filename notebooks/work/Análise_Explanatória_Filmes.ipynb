```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
```

# Removendo dados faltanes

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
df.isnull().sum()
```

```python
df.info()
```

# Mineração de dados

Utilize os dados no intervalo de 2018 a 2023

```python
df_2018_2023  = df[df['Year'] >= 2018]
```

```python
df_2018_2023 = df_2018_2023.drop(columns=['Month', 'Certificate', 'Directors', 'Stars'])
```

```python
df_2018_2023.head()
```

```python
df_2018_2023 = df_2018_2023.replace({'Unknown': np.nan})
df_2018_2023 = df_2018_2023.dropna()
```

```python
df_2018_2023.info()
```

```python
df_2018_2023.head()
```

```python
df_2018_2023 = df_2018_2023.reset_index(drop=True)
```
