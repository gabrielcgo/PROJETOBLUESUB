```python
import pandas as pd
import numpy as np
```

# Removendo dados faltantes

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
df.isnull().sum()
```

```python
df.info()
```

# Mineração de dados

Utilize os dados no intervalo de 2018 a 2023

```python
df_2018_2023  = df[df['Year'] >= 2018]
```

```python
df_2018_2023 = df_2018_2023.drop(columns=['Certificate', 'Directors', 'Stars'])
```

```python
df_2018_2023.head()
```

```python
df_2018_2023 = df_2018_2023.replace({'Unknown': np.nan})
df_2018_2023 = df_2018_2023.dropna()
```

```python
df_2018_2023.info()
```

```python
df_2018_2023.head()
```

```python
df_2018_2023 = df_2018_2023.reset_index(drop=True)
```

```python
df_2018_2023.head()
```

```python
df['Month'].unique()
```

```python
df_2018_2023 = df_2018_2023.replace({'January': '01', 
                                     'February': '02',
                                     'March': '03',
                                     'April': '04',
                                     'May': '05',
                                     'June': '06',
                                     'July': '07',
                                     'August': '08',
                                     'September': '09',
                                     'October': '10',
                                     'November': '11',
                                     'December': '12',
                                     '2008': 'np.nan'})
df_2018_2023 = df_2018_2023.dropna()
```

```python
df_2018_2023.head()
```

```python
df_2018_2023['Month'] = df_2018_2023['Month'].astype(int)

df_2018_2023.info()
```

```python
df_2018_2023.to_csv('movies_2018_2023.csv', index=False)
```
